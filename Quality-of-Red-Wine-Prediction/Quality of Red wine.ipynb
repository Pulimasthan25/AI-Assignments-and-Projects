## Training Test Split

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
DataFrame['quality'] = labelencoder_y.fit_transform(DataFrame['quality'])
```

```python
x = DataFrame.drop('quality', axis = 1)
y = DataFrame['quality']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

##  Model training:

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,mean_squared_log_error
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:10]
```

### Mean Squared Error

```python
meanSquaredError = mean_squared_error(y_test,y_pred)
```

```python
meanSquaredError
```

### Mean Absolute Error

```python
meanAbsoluteError = mean_absolute_error(y_test,y_pred)
```

```python
meanAbsoluteError
```

### R2 Score

```python
r2Score = r2_score(y_test,y_pred)
```

```python
r2Score
```

### Mean Squared Log Error

```python
meanSquaredLogError = mean_squared_log_error(y_test,y_pred)
```
